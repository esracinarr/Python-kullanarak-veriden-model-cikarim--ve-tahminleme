```python
import pandas as pd
```

```python
df = pd.read_csv('vehicle.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.plot(x='compactness', y='hollows_ratio', kind='scatter')
```

```python
df.hist(column='hollows_ratio')
```

```python
from sklearn.preprocessing import PowerTransformer
#log_hollows_ratio = np.log(df_num['hollows_ratio'].values + 1)
pow_transformer = PowerTransformer()
log_hollows_ratio = pow_transformer.fit_transform(df['hollows_ratio'].values.reshape((-1, 1)))
```

```python
df['log_hollows_ratio'] = log_hollows_ratio
```

```python
df.hist(column='log_hollows_ratio')
```

```python
df.corr()
```

```python
#df_features = df.drop(['contraceptive_method_used', 'log_contraceptive_method_used'], axis=1)
#df_features = df.drop(['day', 'month', 'contraceptive_method_used'], axis=1)
df_features = df[['skewness_minor_axis', 'kurtosis_minor_axis', 'kurtosis_major_axis', 'hollows_ratio']]
```

```python
df_features.info()
```

```python
target = df['log_hollows_ratio'].values
```

```python
target.shape
```

```python
from sklearn.model_selection import train_test_split
trainx, testx, trainy, testy = train_test_split(df_features, target, test_size=0.3, random_state=73)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
sc = MinMaxScaler()
trainx_scaled = sc.fit_transform(trainx)
testx_scaled = sc.transform(testx)
```

```python
print(trainx_scaled.shape)
print(testx_scaled.shape)
```

```python
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(trainx_scaled, trainy)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
inv_trainy = pow_transformer.inverse_transform(trainy.reshape(-1, 1))
inv_pred_trainy = pow_transformer.inverse_transform(lm.predict(trainx_scaled).reshape((-1, 1)))
inv_preds = pow_transformer.inverse_transform(lm.predict(testx_scaled).reshape((-1, 1)))
inv_testy = pow_transformer.inverse_transform(testy.reshape((-1, 1)))

print(mean_absolute_error(inv_trainy, inv_pred_trainy))
print(mean_absolute_error(inv_testy, inv_preds))
```

```python
print(r2_score(inv_trainy, inv_pred_trainy))
print(r2_score(inv_testy, inv_preds))
```

```python
svr = SVR(gamma='scale')
svr.fit(trainx_scaled, trainy)
```

```python
inv_trainy = pow_transformer.inverse_transform(trainy.reshape(-1, 1))
inv_pred_trainy = pow_transformer.inverse_transform(svr.predict(trainx_scaled).reshape((-1, 1)))
inv_preds = pow_transformer.inverse_transform(svr.predict(testx_scaled).reshape((-1, 1)))
inv_testy = pow_transformer.inverse_transform(testy.reshape((-1, 1)))

print(mean_absolute_error(inv_trainy, inv_pred_trainy))
print(mean_absolute_error(inv_testy, inv_preds))
```

```python
print(r2_score(inv_trainy, inv_pred_trainy))
print(r2_score(inv_testy, inv_preds))
```

```python
rfr = RandomForestRegressor(n_estimators=100)
rfr.fit(trainx_scaled, trainy)
```

```python
inv_trainy = pow_transformer.inverse_transform(trainy.reshape(-1, 1))
inv_pred_trainy = pow_transformer.inverse_transform(rfr.predict(trainx_scaled).reshape((-1, 1)))
inv_preds = pow_transformer.inverse_transform(rfr.predict(testx_scaled).reshape((-1, 1)))
inv_testy = pow_transformer.inverse_transform(testy.reshape((-1, 1)))


print(mean_absolute_error(inv_trainy, inv_pred_trainy))
print(mean_absolute_error(inv_testy, inv_preds))
```

```python
print(r2_score(inv_trainy, inv_pred_trainy))
print(r2_score(inv_testy, inv_preds))
```
